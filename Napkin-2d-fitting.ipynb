## 2D fit test

```python
result = agf.gaussfitter.gaussfit(image)
```

```python
fig,ax = subplots(figsize=(10,5))
params = result
(height, amplitude, x, y, width_x, width_y, rota) = params

imshow(log(image),cmap=cm.cubehelix)
contour((agf.gaussfitter.twodgaussian(params,shape=image.shape)),cmap=cm.Accent)
       

ax = gca()
text(2, 0.05, """
x : %.1f ,bkg: %.f ,
y : %.1f , peak: .%f ,
width_x : %.1f
width_y : %.1f
rot: %1.f """ %(x, y, height, amplitude, width_x, width_y, rota),
        fontsize=16, horizontalalignment='right',
        verticalalignment='bottom', transform=ax.transAxes)

colorbar()


```
