```python
def shade(on_off, ax):
    for on, off in zip(on_off[:-1:2], on_off[1::2]):
            on = on + data.index.values.min()
            off = off + data.index.values.min()
            ax.add_patch(Rectangle((on, -1), off - on, 2, facecolor="grey"))
```

```python
name = 'InAs0799time3'
data, parameters_dict, on_off, offset_list = loadme(name)
fig_1, ax_1 = do_diff_derivatives(
    data, parameters_dict, 'TW', name, False, False, True, True,figsize=(20,8))

ax_1.set_xlim(300,800)
ax_1.set_xlabel('Time(s)')
```

```python
x = data.index
y = data.Int_WZ
xs_WZ, ys_WZ, d_ys_WZ, dd_ys_WZ = spline_do(
    x, y, [ax_1], n=3, s=parameters_dict['WZ'], k=3, plot=False)
```

```python
x = data.index
y_TW = data.Int_TW
xs_TW, ys_TW, d_ys_TW, dd_ys_TW = spline_do(
    x, y_TW, [ax_1], n=3, s=parameters_dict['TW'], k=3, plot=False)
```

WZ

```python
fig,ax = subplots(figsize=(15,10))
ax.plot(xs_WZ,d_ys_WZ,label='spline')
for mean in [13,21]:
    for median in [10,41]:
        plot(data.index, pd.rolling_mean(pd.rolling_median(data.Int_WZ,median).diff(),mean),'-',label = '''moving avg window:{0},
        moving median window:{1}'''.format(mean,median) )
ax.legend()
```

```python
ax.set_xlim(50,400)
ax.set_ylim(-.01,.04)
shade(on_off,ax)
fig
```

TW

```python
fig,ax = subplots(figsize=(15,10))
ax.plot(xs_TW,d_ys_TW,label='spline')
for mean in [13,21]:
    for median in [10,41]:
        plot(data.index, pd.rolling_mean(pd.rolling_median(data.Int_TW,median).diff(),mean),'-',label = '''moving avg window:{0},
        moving median window:{1}'''.format(mean,median) )
ax.legend()
```

```python
ax.set_xlim(50,400)
ax.set_ylim(-.01,1)
shade(on_off,ax)
fig
```
