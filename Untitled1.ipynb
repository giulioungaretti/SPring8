Explaining the decrease in the intesities of TW, WZ by increase of ZB

```python
fig, ax = subplots(figsize=(15,10))

ax2  = ax.twinx()
ax3  = ax.twinx()

ax.plot(data.index, pd.rolling_median(data.Int_TW,1),'.',label='data TW',color ='#D95B43')
ax.plot(xs_TW,ys_TW,'-',label='Spline TW',color='#ECD078')

shade(on_off, ax)
ax2.plot(data.index, pd.rolling_median(data.Int_WZ,1),'.',label='data WZ',color='#C02942')
ax2.plot(xs_WZ,ys_WZ,'-',label='Spline WZ',color='#542437')

ax3.plot(data.index, pd.rolling_median(data.Int_ZB,1),'.',label='data ZB')
ax3.plot(xs_ZB,ys_ZB,'-',label='Spline ZB',color='#53777A')


ax.set_xlim(100,400)
ax.set_ylim(250,285)
ax2.set_ylim(12,15)
ax3.set_ylim(12,15)
update_legend(fig)
```
