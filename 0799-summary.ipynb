```python
time = numpy.concatenate([x1,x2,x3,x4])
int_wz = numpy.concatenate([y1,y2,y3,y4])
```

```python
step = 50
delta_h =  int_wz[step::step]- int_wz[:-step:step]
h = np.cumsum(delta_h)
```

```python
# fig, ax2 = subplots(figsize=(15,5))
ax2= ax.twinx()
lefts = time[::50][:-1]
    
ax2.bar(lefts,h ,  align='center', width=30,color='black')

ax2.bar(lefts,h, width=10,  align='center', color='grey')

ax2.set_frame_on(False)
ax2.get_xaxis().tick_bottom()
ax2.get_yaxis().tick_left()
[i.grid('off') for i in fig.get_axes()]
# ax2.axes.get_yaxis().set_visible(False)
ax2.set_xlim(0,3500)
ax2.set_yticklabels([])

xmin, xmax = ax2.get_xaxis().get_view_interval()
ymin, ymax = ax2.get_yaxis().get_view_interval()
ax2.add_artist(Line2D((xmin, xmax), (ymin, ymin), color='black', linewidth=2))
ax2.set_xlabel('Time (s)')
ax2.set_ylabel(' NW lenght (a.u.)')

for l,height in zip(lefts,h):
    circle = mpatches.Ellipse([l,height],25,4 , color = 'y')
    ax2.add_patch(circle)
    
[i.set_alpha(0) for i in ax.get_lines()]
[i.grid('off') for i in fig.get_axes()]
ax.set_frame_on(False)
ax.get_xaxis().tick_bottom()

ax.axes.get_yaxis().set_visible(False)

###  add x line
```
