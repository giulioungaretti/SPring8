```python
class SPring8_image(object):

    '''
    image class that reads from decetor, ccd, pilatus

    ---
    list of attributes 
    file_name =  name of the file to read (string)
    folder = is the path of the files to be read, default none
    time = time stamp in second
    ROI = the region of interest selected
    '''


    def __init__(self, file_name,folder=None, ROI=None):

        if not folder: self.file_name = file_name
        if folder: self.file_name = str(folder)+str(file_name)
        self.ROI = ROI

    def read_detector(self, name=1):
        '''
        Read the detector image and return a dictionary with the information contained
        in the header and the image as a numpy array.
        name can be 1 for ccd or 2 for pilatus
        TO-Do add angles from the diffractometers
        '''
        self.header_dict = {}
        if name == 1:
            byte = 2048
            res = 512
            user_dtype = uint16
        if name == 2:
            byte = 4096
            res = [487, 195]  # rect
            user_dtype = int32
        # if you only wanted to read 512 bytes, do .read(512)
        in_file = open(self.file_name, 'rb')
        header = in_file.read(byte)
        image = in_file.read()  # read the rest of image
        in_file.close()  # save memory
        # convert the  binary image to integers
        if name == 1:
            self.img = np.fromstring(
                image, dtype=user_dtype, count=res * res).reshape(res, res)
        if name == 2:
            self.img = np.fromstring(
                image, dtype=user_dtype).reshape(res[1], res[0])
        self.img = median_filter(self.img, 3)  # perform median filter 3x3
        header = header.split('\n')
        for i in range(len(header)):
            if '#L' in header[i]:
                for k, l in zip(filter(None, header[i][2:].split(' ')),  # not robust way to filter out the comment on the first line
                                filter(None, header[i + 1].split(' '))):
                    self.header_dict[k] = l
            if '#V1' in header[i]:
                # first presssure saved
                self.NIG = float(header[i].split(' ')[3])
                self.Tc = float(header[i].split(' ')[6])  # silly detail
                self.Tpyro = float(header[i].split(' ')[9])  # silly detail
            if '#C3' in header[i]:
                self.c3 = header[i]
            if '#V2' in header[i]:
                if header[i ].split(' ')[0] == 'shutter':
                    # save iformation on the shutters 
                    self.Ga =  int(header[i ].split(' ')[3])
                    self.In =  int(header[i ].split(' ')[4])
                    self.Au =  int(header[i ].split(' ')[5])
                    self.Sb = int(header[i ].split(' ')[6])
                    self.As =  int(header[i ].split(' ')[7])
                # sometimes the reading just fucks up
                if header[i ].split(' ')[2] == 'graphtech':
                    self.NIG = float(header[i ].split(' ')[3])
                    self.Tpyro = float(header[i].split(' ')[4])  # silly detail
                   # self.NIG2 = float(header[i ].split(' ')[7]) # what is this pressure I don't know!
        try:  
            #### TO- DO add more attributes
            
            # time of the exposure in seconds
            self.time = float(self.header_dict['Epoch'])
            # some measure  of the intesnity of the incoming beam
            # use to normalize
            self.monitor = float(self.header_dict['Monitor'])

        except Exception, e:
            print e

    def show(self,log=False):
        'sowh detecotr image, optioanlly in log scale'
        fig, ax = subplots()
        ax.grid(color='white')
        if log : imshow(np.log(self.img), cmap=cm.Oranges)
        if not log :  imshow(self.img, cmap=cm.Oranges)
        plt.colorbar()
        return fig,ax
        
        
    def select_ROI(self, img, display_plot=False, ROI=None,log=False):
        '''
        takes single image as input convert to gray scale and ask to select rectangular roi
        '''
        self.ROI = ROI
        try:
            ioff()
        except Exception, e:
            print e
        fig, ax = subplots()
        ax.grid(color='black')
        if type(self.ROI) == list:
            x_1, x_2, y_1, y_2 = self.ROI
            ax.grid(color='black')
            img_intrest = img[y_1:y_2, x_1:x_2]
            if display_plot:
                imshow(img_intrest)  # (yy,xx)
            if display_plot:
                display(fig)
        else:
            if display_plot:
                if log : imshow(np.log(self.img), cmap=cm.Oranges)
                if not log :  imshow(self.img, cmap=cm.Oranges)
            if display_plot:
                try:
                    display(fig)
                except Exception, e:
                    print e
            cond = False
            while not cond:
                x_1 = float(raw_input('x_1:'))
                x_2 = float(raw_input('x_2:'))
                y_1 = float(raw_input('y_1:'))
                y_2 = float(raw_input('y_2:'))
                img_intrest = img[y_1:y_2, x_1:x_2]
                if display_plot:
                     if log : imshow(np.log(img_intrest), cmap=cm.Oranges)
                     if not log :  imshow(img_intrest, cmap=cm.Oranges)
                if display_plot:
                    try:
                        display(fig)
                    except Exception, e:
                        print e
                check = str(raw_input('return if ok,space to redo'))
                if len(check) < 1:
                    cond = True
        try:
            close('all')
            ion()
        except Exception, e:
            print e

        self.ROI = [x_1, x_2, y_1, y_2]
        self.img_roi = img_intrest

    def analyze_img(self, std=3, fit=False, plot_me=False, details=False):
        '''
        analyze the image and gives the result black.
        Fitting with Gaussian and showing the cuts, and the fits is controlled via the 
        optional boolean paramters fit, plot_mem and details
        '''
        try:
            ioff()
        except Exception, e:
            print e
        results = []
        if type(self.ROI) == list:
            x_1, x_2, y_1, y_2 = self.ROI
            max_int_pos = where(self.img_roi >= self.img_roi.max())
            y_cut = max_int_pos[1]
            x_cut = max_int_pos[0]
            img_crop_x = self.img_roi[x_cut][0]
            img_crop_y = self.img_roi[:, y_cut].flatten()
            xx = linspace(y_1, y_2, len(img_crop_x))
            xy = linspace(x_1, x_2, len(img_crop_y))

            if plot_me:
                fig_update, ax_update = subplots()
                imshow(self.img_roi, interpolation='nearest', cmap=cm.Oranges)
                ax_update.axhline(x_cut, color='red', ls='--')
                ax_update.axvline(y_cut, color='black', ls='--')
                ax_update.set_title(str(image)[-9:-7] + 'minutes of growth')
                display(fig_update)
            try:
                y = img_crop_x
                amplitude = img_crop_x.max()
                mean_x = xx[where(img_crop_x == img_crop_x.max())]
                if fit:
                    poptx, pcov = curve_fit(
                        gauss_function, xx, y, p0=[amplitude, mean_x, std])
                # fit y
                y = img_crop_y
                amplitude = img_crop_y.max()
                mean_y = xy[where(img_crop_y == img_crop_y.max())]
                if fit:
                    popty, pcov = curve_fit(
                        gauss_function, xy, y, p0=[amplitude, mean_y, std])
                    results.append([poptx, popty])
                    if details:
                        fig_det, [ax_det1, ax_det2] = subplots(2, 1)
                        ax_det1.plot(xx, img_crop_x, 'o')
                        ax_det1.plot(
                            xx, gauss_function(xx, amplitude, mean_x, std))
                        ax_det2.plot(xy, img_crop_y, 'o')
                        ax_det2.plot(
                            xy, gauss_function(xy, amplitude, mean_y, std))
                        display(fig_det)
            except Exception as e:
                print 'fit problem', e
                poptx = None
                popty = None

        else:
            print 'not a list, aborting'

        try:
            if fit:
                self.max_pos_x = poptx[1]
                self.max_pos_y = popty[1]
                self.max_x = poptx[0]
                self.width_x = poptx[2]
                self.max_y = popty[0]
                self.width_y = popty[2]
                self.int_intensity = np.sum(self.img[self.max_pos_x - 10:self.max_pos_x + 10, self.max_pos_y - 10:self.max_pos_y + 10])
                self.max_intensity = (self.max_y + self.max_x) / 2
            else:
                self.max_pos_x = center_of_mass(self.img_roi)[0]
                self.max_pos_y = center_of_mass(self.img_roi)[1]
                self.int_intensity = np.sum(self.img_roi)

        except Exception, e:
            print e

        try:
            close('all')
            ion()
        except Exception, e:
            print e
```

```python
def get_roi(i,kind,log=False):
    temp_image = SPring8_image(i)
    temp_image.read_detector(name=kind)
    temp_image.select_ROI(temp_image.img,True,log=log)
    return temp_image.ROI
```

```python
def parse_sample(files,samples,roi):
    '''
    list of filenames and retrun result of batch analyis
    '''
    
    try:
        ion()
    except Exception as e:        
        print e
    sample = [] 
    for i in files:
        try:
           
            temp_image = SPring8_image(i)
            temp_image.read_detector(name=1)
            temp_image.select_ROI(temp_image.img,ROI=roi)
            temp_image.analyze_img()
            sample.append(temp_image)
        except Exception as e:
            print 'error'+str(e)+'file:'+i
    samples[str(roi)] = sample
    return samples
```
